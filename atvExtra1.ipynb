```python
import math
def gerador_quantizacao():
  # Solicita o número de níveis de quantização ao usuário
  n_niveis = int(input("Insira o número de níveis de quantização: "))

  # Calcula o número de bits necessários (arredonda para cima)
  n_bits = math.ceil(math.log2(n_niveis))

  # Exibe os resultados calculados
  print(f"\n--- Resultado ---")
  print(f"Niveis desejados: {n_niveis}")
  print(f"Bits necessários: {n_bits}")
  print(f"Capacidade real com esses bits: {2**n_bits} combinações")
  print(f"---------------------\n")

  # Exibe as combinações binárias para o número de bits
  print(f"Combinações de {n_bits} bits: ")
  for i in range(2**n_bits):
      combinacao = f"{i:0{n_bits}b}"
      print(f"Nivel {i:3}: {combinacao}")

# Executa a função principal quando o script é iniciado
if __name__== "__main__":
    gerador_quantizacao()
```

```python
def binario_para_decimal(binario_str):
    decimal = 0
    potencia = 0

    # Inverte a string binária para facilitar o cálculo (do bit menos significativo ao mais significativo)
    binario_str = binario_str[::-1]

    # Itera sobre cada dígito (bit) da string binária invertida
    for digito in binario_str:
        # Se o dígito for '1', adiciona 2 elevado à potência atual ao valor decimal
        if digito == '1':
            decimal += 2**potencia
        # Se o dígito não for '0' nem '1', é uma entrada inválida
        elif digito != '0':
            return "Entrada inválida: a string deve conter apenas '0' ou '1'."
        # Aumenta a potência para o próximo dígito
        potencia += 1
    # Retorna o valor decimal calculado
    return decimal

# Solicita ao usuário que digite um número binário
binario_input = input("Digite um número binário: ")
# Chama a função para converter o binário para decimal
resultado = binario_para_decimal(binario_input)
# Imprime o resultado da conversão
print(f"O número decimal de {binario_input} é: {resultado}")
```

```python
# Essa função converte um número decimal para binário.
def decimal_binario(n):
    # Se o número for 0, o binário é "0".
    if (n) == 0:
        return "0"

    # Prepara uma string vazia para guardar os bits.
    bits = ""
    # Continua dividindo o número por 2 enquanto ele for maior que 0.
    while n > 0:
      # Pega o resto da divisão por 2 (que será 0 ou 1).
      resto = n % 2
      # Adiciona o resto no início da string 'bits'.
      bits = str(resto) + bits
      # Divide o número por 2 e pega só a parte inteira.
      n = n // 2

    # Devolve a string binária formada.
    return bits

# Pede ao usuário para digitar um número decimal.
numero = int(input("Digite um número decimal: "))
# Chama a função para converter o número e imprime o resultado.
print(f"0 binário de {numero} é: {decimal_binario(numero)}")
```

```python
# Função para converter um número decimal (com parte inteira e fracionária) para binário
def decimal_para_binario_completo(decimal_num, precisao=10):
    if not isinstance(decimal_num, (int, float)):
        return "Entrada inválida. Digite um número decimal."

    parte_inteira = int(decimal_num)
    parte_fracionaria = decimal_num - parte_inteira

    # Convertendo a parte inteira (método de divisão sucessiva por 2)
    if parte_inteira == 0:
        binario_inteiro = "0"
    else:
        binario_inteiro = ""
        temp_inteira = parte_inteira
        while temp_inteira > 0:
            binario_inteiro = str(temp_inteira % 2) + binario_inteiro
            temp_inteira //= 2

    # Convertendo a parte fracionária (método de multiplicação sucessiva por 2)
    if parte_fracionaria == 0:
        binario_frac = ""
    else:
        binario_frac = "."
        for _ in range(precisao):
            parte_fracionaria *= 2
            if parte_fracionaria >= 1:
                binario_frac += "1"
                parte_fracionaria -= 1
            else:
                binario_frac += "0"
            if parte_fracionaria == 0: # Se a parte fracionária se tornar 0, podemos parar
                break

    # Combinando a parte inteira e a parte fracionária binárias
    if binario_frac:
        return binario_inteiro + binario_frac
    else:
        return binario_inteiro

# Solicita ao usuário um número decimal
num_decimal = float(input("Digite um número decimal (ex: 10.625 ou 0.75): "))

# Solicita a precisão desejada para a parte fracionária binária
precisao_input = int(input("Digite a precisão para a parte fracionária (número de casas decimais binárias, ex: 10): "))

# Realiza a conversão e imprime o resultado
resultado_binario = decimal_para_binario_completo(num_decimal, precisao_input)
print(f"O número binário de {num_decimal} é: {resultado_binario}")
```
